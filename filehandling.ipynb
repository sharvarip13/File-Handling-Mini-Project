```python
f=open("NewStudent.txt","r")
roll_no=input("Enter Roll No to search:")
for line in f:
    if roll_no in line:
        print("Record found:=",line)
f.close()
```

```python
roll_no=input("Enter Roll No:")
new_marks=input("Enter New Marks:")
f=open("NewStudent.txt","r")
lines=f.readlines()
f.close()
f=open("NewStudent.txt","w")
for line in lines:
    data=line.strip().split()
    if data[1]==roll_no:
        line=data[0]+","+data[1]+","+new_marks+"\n"
    f.write(line)
f.close()
print("Marks are updated")
```

```python
key=input("Enter Name or Roll No:")
f=open("NewStudent.txt","r")
lines=f.readlines()
f.close()
f=open("NewStudent.txt","w")
for line in lines:
    data=line.strip().split(" ")
    if data[1]!=key and data[1]!=key:
        f.write(line)
f.close()
print("Record is deleted")
```

```python
f=open("NewStudent.txt","r")
records=[]
for line in f:
    records.append(line.strip().split())
f.close()
records.sort(key=lambda x:int(x[2]),reverse=True)
f=open("Top3.txt","w")
for i in records[:3]:
    f.write(",".join(i)+"\n")
f.close()
print("Top 3 exported")
```

```python
f=open("NewStudent.txt","r")
above_90=0
below_35=0
for line in f:
    data=line.strip().split()
    if int(data[2])>90:
        above_90+=1
    elif int(data[2])<35:
        below_35+=1
f.close()
print("Students scoring above 90:",above_90)
print("Students scoring below 35:",below_35)
```
